```python
import mecab_ko as MeCab
import ckip_transformers
from ckip_transformers.nlp import CkipWordSegmenter
```

```python
path_to_chinese_train_data = "./data/zh_train.txt"
path_to_chinese_train_output = "./zh_train_ckip_output.txt"

path_to_chinese_test_data = "./data/zh_test.txt"
path_to_chinese_test_output = "./zh_test_ckip_output.txt"

path_to_korean_train_data = "./data/ko_train.txt"
path_to_korean_train_output = "./data/ko_train_mecab_output.txt"

path_to_korean_test_data = "./data/ko_test.txt"
path_to_korean_test_output = "./data/ko_test_mecab_output.txt"
```

Chinese segmentation with CKIP Transformers:

```python
ws_driver = CkipWordSegmenter(model="bert-base", device=0)
```

```python
with open(path_to_chinese_train_data) as file:
    lines = file.readlines()
    lines = [line.rstrip() for line in lines]
```

```python
segged=ws_driver(lines)

with open(path_to_chinese_train_output, "w") as text_file:
    text_file.write("%s" % '\n'.join([' '.join(p) for p in segged]))
```

```python
with open(path_to_chinese_test_data) as file:
    lines = file.readlines()
    lines = [line.rstrip() for line in lines]
```

```python
segged=ws_driver(lines)

with open(path_to_chinese_test_output, "w") as text_file:
    text_file.write("%s" % '\n'.join([' '.join(p) for p in segged]))
```

Korean segmentation with MeCab-ko:

```python
tagger = MeCab.Tagger("-Owakati")
```

```python
with open(path_to_korean_train_data) as file:
    lines = file.readlines()
    lines = [line.rstrip() for line in lines]
```

```python
with open(path_to_korean_train_output, "w") as text_file:
  for i in lines:
    text_file.write(tagger.parse(i))
```

```python
with open(path_to_korean_test_data) as file:
    lines = file.readlines()
    lines = [line.rstrip() for line in lines]
```

```python
with open(path_to_korean_test_output, "w") as text_file:
  for i in lines:
    text_file.write(tagger.parse(i))
```

Char-based segmentation:

```python
path_to_chinese_train_char_output = "./zh_train_char_output.txt"
path_to_chinese_test_char_output = "./zh_test_char_output.txt"
path_to_korean_train_char_output = "./ko_train_char_output.txt"
path_to_korean_test_char_output = "./ko_test_char_output.txt"

char_input_paths = [path_to_chinese_train_data,
                    path_to_chinese_test_data,
                    path_to_korean_train_data,
                    path_to_korean_test_data]

char_output_paths = [path_to_chinese_train_char_output, 
              path_to_chinese_test_char_output, 
              path_to_korean_train_char_output,
              path_to_korean_test_char_output]
```

```python
for char_path in zip(char_input_paths, char_output_paths):
  with open(char_path[0]) as file:
    lines = file.readlines()
    lines = [line.replace(" ", "").rstrip() for line in lines]

  with open(char_path[1], "w") as text_file:
    for i in lines:
      text_file.write(" ".join(list(", ".join(map(str, i.split()))))+'\n')
```
